```python
import pandas as pd

from sql_connection import table_df
```

```python
df = table_df('customer', 'new_customer_profile')
```

```python
df.head()
```

```python
def data_profile(df):
    null_count = df.isna().sum()
    null_count_df = pd.DataFrame(null_count, columns=['null_count']).transpose()
    describe = df.describe()
    return pd.concat([null_count_df, describe], axis=0)
```

```python
my_data_profile = data_profile(df)
```
