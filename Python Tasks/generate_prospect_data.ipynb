```python
import pandas as pd
import random
import numpy as np

random.seed(42)

def generate_prospect_data(num_records=5000000):

    first_name = [f"Robert{i+1}" for i in range(num_records)]
    last_name = ["Murray"] * num_records
    email = [f"{fn.lower()}.{ln.lower()}@mail.com" for fn, ln in zip(first_name, last_name)]
    addressP_ward_no = [str(random.randint(1, 35)) for _ in range(num_records)]
    addressT_ward_no = addressP_ward_no
    empty_column = [np.nan] * num_records

    data = {
        "salutation": [random.choice(["Mr.", "Mrs."]) for _ in range(num_records)],
        "firstname": first_name,
        "middlename": empty_column,
        "lastname": last_name,
        "gender": [random.choice(["M", "F"]) for _ in range(num_records)],
        "email": email,
        "phone_number": empty_column,
        "mobile_number": empty_column,
        "addressP_street_address": [f"Street{i+1}" for i in range(num_records)],
        "addressP_ward_no": addressP_ward_no,
        "addressP_municipality": ["Lalitpur"] * num_records,
        "addressP_city": ["Lalitpur"] * num_records,
        "addressP_line": empty_column,
        "addressP_district": ["Lalitpur"] * num_records,
        "addressP_state": ["Bagmati"] * num_records,
        "addressP_country": ["Nepal"] * num_records,
        "addressT_street_address": [f"Street{i+1}" for i in range(num_records)],
        "addressT_ward_no": addressT_ward_no,
        "addressT_municipality": ["Lalitpur"] * num_records,
        "addressT_city": ["Lalitpur"] * num_records,
        "addressT_line": empty_column,
        "addressT_district": ["Lalitpur"] * num_records,
        "addressT_state": ["Bagmati"] * num_records,
        "addressT_country": ["Nepal"] * num_records,
        "dob": empty_column,
        "marital_status": [random.choice(["S", "M", "D"]) for _ in range(num_records)],
        "email_disable": [random.choice([0, 1]) for _ in range(num_records)],
        "phone_disable": [random.choice([0, 1]) for _ in range(num_records)],
        "sms_disable": [random.choice([0, 1]) for _ in range(num_records)],
        "viber_disable": [random.choice([0, 1]) for _ in range(num_records)],
        "push_notification_disable": [random.choice([0, 1]) for _ in range(num_records)],
        "active_flag": [1] * num_records,
        "created_by": ["Application"] * num_records,
        "description": [f"Description{i+1}" for i in range(num_records)],
        "prospect_source": [f"Source{i+1}" for i in range(num_records)],
    }
    return pd.DataFrame(data)
```

```python

prospect_data = generate_prospect_data()
```
